```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import norm
from sklearn.preprocessing import StandardScaler
from scipy import stats
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
path = 'data/sm.ms.solutions-cpuv3'
cpu = pd.read_csv(path, names=['id', 'x', 'y', 'z'], delimiter=' ', na_values=['XXX'], thousands=',', skiprows=0)
#cpu.drop(['id'], axis=1)
#del cpu.id
cpu.drop('id', axis=1, inplace=True)
cpu.describe()
```

```python
path = 'data/sm.ms.solutions-gpuv3'
gpu = pd.read_csv(path, names=['id', 'x', 'y', 'z'], delimiter=' ', na_values=['XXX'], thousands=',', skiprows=0)
#gpu.drop(['id'], axis=1)
#del gpu.id
gpu.drop('id', axis=1, inplace=True)
gpu.describe()
```

```python
sns.set()
cols = ['x', 'y', 'z']
sns.pairplot(cpu[cols], size = 2.5)
plt.show();
```

```python
sns.set()
cols = ['x', 'y', 'z']
sns.pairplot(gpu[cols], size = 2.5)
plt.show();
```

```python
g1 = sns.PairGrid(cpu)
p1 = g1.map(plt.scatter)
p1.savefig("cpu_scatter.png")
```

```python
g2 = sns.PairGrid(cpu)
p2 = g2.map(plt.plot)
p2.savefig("cpu_plot.png")
```

```python
g3 = sns.PairGrid(gpu)
p3 = g3.map(plt.scatter)
p3.savefig("gpu_scatter.png")
```

```python
g4 = sns.PairGrid(gpu)
p4 = g4.map(plt.plot)
p4.savefig("gpu_plot.png")
```
